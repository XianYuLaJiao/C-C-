```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.preprocessing import MinMaxScaler
from keras.callbacks import TensorBoard
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout,GRU
# from keras.optimizers import SGD# 随机梯度下降法
```

```python
dataset = pd.read_excel(r'C:\Users\17180\Desktop\Code\Study-Cave\磨损量数据.xlsx')
```

```python
dataset[0].plot(figsize=[25,8], legend=True,color='g')
plt.title('MSL')
plt.legend(["SH500"])
plt.show()
```

```python
train_set = dataset[:500].iloc[:,1:2]
test_set = dataset[500:].iloc[:,1:2]
```

```python
train_set
```

```python
sc = MinMaxScaler(feature_range=[0,1])# 最大最小值归一化
train_set_scaled = sc.fit_transform(train_set)
print(train_set_scaled.shape)
train_set_scaled
```

```python
x_train = []
y_train = []
for i in range(60, 500):
    x_train.append(train_set_scaled[i-60:i,0])
    y_train.append(train_set_scaled[i,0])
         
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
print(y_train)
x_train[0:2]
```

```python
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
x_train
```

```python
model = Sequential()
#LSTM 第一层
model.add(LSTM(128,return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))

# 第二层
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(0.2))

#第三层
model.add(LSTM(128))
model.add(Dropout(0.2))

# Dense层
model.add(Dense(units=1))
y_train
# 模型编译
model.compile(optimizer='adadelta',loss='mse')

# 训练模型
model.fit(x_train, y_train, epochs=40, batch_size=72)
```
